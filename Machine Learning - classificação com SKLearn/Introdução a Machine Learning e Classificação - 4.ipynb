```python
from datetime import datetime

import pandas as pd
import graphviz
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.dummy import DummyClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.tree import DecisionTreeClassifier, export_graphviz


uri = "https://gist.githubusercontent.com/guilhermesilveira/4d1d4a16ccbf6ea4e0a64a38a24ec884/raw/afd05cb0c796d18f3f5a6537053ded308ba94bf7/car-prices.csv"
data = pd.read_csv(uri, index_col=0)
data.head()
```

```python
data['sold'] = data['sold'].map({'yes': 1, 'no': 0})
data.head()
```

```python
data['model_age'] = datetime.today().year - data['model_year']
data.head()
```

```python
data['total_km'] = data['model_age'] * data['mileage_per_year'] * 1.60934
data.head()
```

```python
del data['mileage_per_year']
del data['model_year']

data.head()
```

```python
X = data[['price', 'model_age', 'total_km']]
Y = data['sold']
```

```python
SEED = 5
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y, random_state=SEED, test_size=0.25)

clf_linear_svc = LinearSVC(random_state=SEED)
clf_linear_svc.fit(X_train, Y_train)

accuracy = clf_linear_svc.score(X_test, Y_test)
print('Precisão Linear SVC: %.2f%%' % (accuracy * 100))
```

```python
baseline_stratified = DummyClassifier(random_state=SEED)
baseline_stratified.fit(X_train, Y_train)

accuracy = baseline_stratified.score(X_test, Y_test)
print('Precisão Stratified: %.2f%%' % (accuracy * 100))
```

```python
baseline_most_frequent = DummyClassifier(strategy='most_frequent', random_state=SEED)
baseline_most_frequent.fit(X_train, Y_train)

accuracy = baseline_most_frequent.score(X_test, Y_test)
print('Precisão Most Frequent: %.2f%%' % (accuracy * 100))
```

```python
raw_X_train, raw_X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y, random_state=SEED, test_size=0.25)

scaler = StandardScaler()
X_train = scaler.fit_transform(raw_X_train)
X_test = scaler.transform(raw_X_test)

clf_svc = SVC(gamma='auto', random_state=SEED)
clf_svc.fit(X_train, Y_train)

accuracy = clf_svc.score(X_test, Y_test)
print('Precisão SVC: %.2f%%' % (accuracy * 100))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y, random_state=SEED, test_size=0.25)

clf_decision_tree = DecisionTreeClassifier(random_state=SEED, max_depth=3)
clf_decision_tree.fit(X_train, Y_train)

accuracy = clf_decision_tree.score(X_test, Y_test)
print('Precisão Decision Tree: %.2f%%' % (accuracy * 100))
```

```python
dot_data = export_graphviz(clf_decision_tree, out_file=None, feature_names=X.columns, filled=True, rounded=True, class_names=['não', 'sim'])
graph = graphviz.Source(dot_data)
graph
```
